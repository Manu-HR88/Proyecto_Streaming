### Transformación de los datos de rating

```python
# Unimos los dataframes de ratings en uno solo.
df_ratings = pd.concat([rating1,rating2,rating3,rating4,rating5,rating6,rating7,rating8], axis = 0, ignore_index=True)

# Comprobamos el resultado
df_ratings
```

Modificamos la columna 'timestamp' que contiene el formato de fecha en UNIX para convertirlo a formato AAAA-MM-DD

```python
# Utilizamos el método to_datetime para obtener la fecha con horas, minutos y segundos.
df_ratings['timestamp'] = pd.to_datetime(df_ratings['timestamp'], unit='s')

# Extraemos solamente la fecha como la necesitamos.
df_ratings['timestamp'] = df_ratings['timestamp'].dt.strftime('%Y-%m-%d')

#Comprobamos el resultado
df_ratings
```

```python
# Exportamos a archivo parquet para utilizar en el modelo de machine learning.
table = pa.Table.from_pandas(df_ratings)
pq.write_table(table, 'datasets/ratings.parquet')
```

```python
df_ratings.to_csv('datasets/ratings.csv')
```

```python
# Verificamos el valor minimo y máximo de las fechas transformadas.

print('El valor más antiguo de las fechas es: ', df_ratings['timestamp'].min())
print('El valor más reciente de las fechas es: ', df_ratings['timestamp'].max())
```

### Transformaciones

1. Generar un campo id. En este caso lo generaremos tomando como base el la forma del id de los archivos "ratings" que consiste en las tres primeras letras del nombre de la plataforma y el número de show_id de las bases.

```python
# Generación del campo id de la base de Amazon Prime Video
df_amzn.insert(loc=0, column = 'movieId', value= ('a' + df_amzn['show_id']))

# Generación del campo 'platform' que utilizaremos para la creación de las clases para los filtros de las funciones
df_amzn.insert(loc=1, column = 'platform', value= 'amazon')

# Eliminamos la columna 'show_id' que ya no será util
df_amzn = df_amzn.drop('show_id', axis=1)

# Comprobamos el cambio realizado
df_amzn.head()
```

```python
# Generación del campo id de la base de Disney
df_disn.insert(loc=0, column = 'movieId', value= ('d' + df_disn['show_id']))

# Generación del campo 'platform' que utilizaremos para la creación de las clases para los filtros de las funciones
df_disn.insert(loc=1, column = 'platform', value= 'disney')

# Eliminamos la columna 'show_id' que ya no será util
df_disn = df_disn.drop('show_id', axis=1)

# Comprobamos el cambio realizado
df_disn.head()
```

```python
# Generación del campo id de la base de Hulu
df_hulu.insert(loc=0, column = 'movieId', value= ('h' + df_hulu['show_id']))

# Generación del campo 'platform' que utilizaremos para la creación de las clases para los filtros de las funciones
df_hulu.insert(loc=1, column = 'platform', value= 'hulu')

# Eliminamos la columna 'show_id' que ya no será util
df_hulu = df_hulu.drop('show_id', axis=1)

# Comprobamos el cambio realizado
df_hulu.head()
```

```python
# Generación del campo id de la base de Netflix
df_nflx.insert(loc=0, column = 'movieId', value= ('n' + df_nflx['show_id']))

# Generación del campo 'platform' que utilizaremos para la creación de las clases para los filtros de las funciones
df_nflx.insert(loc=1, column = 'platform', value= 'netflix')

# Eliminamos la columna 'show_id' que ya no será util
df_nflx = df_nflx.drop('show_id', axis=1)

# Comprobamos el cambio realizado
df_nflx.head()
```

Ahora que tenemos un id único para identificar cada registro de cada dataframe, podemos proceder a juntarlos para realizar las demás transformaciones

```python
df_films = pd.concat([df_amzn,df_disn,df_hulu,df_nflx], axis = 0, ignore_index=True)
df_films.tail()
```

```python
df_films.info()
```

2. Reemplazar los valores nulos del campo rating con una "G"

```python
#Como la columna 'rating' no hace referencia a la puntuación sino a la clasificación de la película, haremos el cambio de nombre
df_films = df_films.rename({'rating':'clasificacion'}, axis = 1)
```

```python
df_films
```

```python
# Se utiliza el método fillna para substituir los valores nulos por la "G"
df_films['clasificacion'] = df_films['clasificacion'].fillna('G')
```

```python
#Comprobamos que la columna 'clasificacion' ya no tenga nulos.
df_films.info()
```

3. Cambiar el formato de fecha de la columna date_added (AAAA/mm/dd)

```python
# Primero eliminamos los espacios en blanco anteriores y posteriores a la fecha.
df_films['date_added'] = df_films['date_added'].str.strip()

# Una vez limpios procedemos a realizar el cambio con el método "to_datetime"
df_films['date_added'] = pd.to_datetime(df_films["date_added"], format='%B %d, %Y')

# Comprobamos el cambio realizado.
df_films.head()
```

```python
print('la fecha minima: ', df_films['date_added'].min())
print('la fecha maxima: ', df_films['date_added'].max())
```

4. Cambiar los campos de texto a minúsculas.

```python
# Definimos cuáles van a ser las columnas de texto que queremos cambiar.
# 'type', 'title', 'director', 'cast', 'country', 'rating', 'duration', 'listed_in', 'description'

df_films['type'] = df_films['type'].str.lower()
df_films['title'] = df_films['title'].str.lower()
df_films['director'] = df_films['director'].str.lower()
df_films['cast'] = df_films['cast'].str.lower()
df_films['country'] = df_films['country'].str.lower()
df_films['clasificacion'] = df_films['clasificacion'].str.lower()
df_films['duration'] = df_films['duration'].str.lower()
df_films['listed_in'] = df_films['listed_in'].str.lower()
df_films['description'] = df_films['description'].str.lower()
```

5. Cambiar la columna 'duration' en dos campos: 'duration_int' y 'duration_type'

```python
# Hacemos uso del método split() para separar por espacio el valor de 'duration'
df_films[['duration_int', 'duration_type']] = df_films['duration'].str.split(' ', n=1, expand=True)

# Para poder cambiar la columna 'duration_int' a entero, debemos cambiar el valor de los nulos a "0" que es un valor entero
df_films['duration_int'] = df_films['duration_int'].fillna(0)

# Cambiamos el tipo de la columna 'duration_int' de object a int
df_films['duration_int'] = df_films['duration_int'].astype("int64")

# Cambiamos el valor 'seasons' a 'season' en la columna duration_type
df_films['duration_type'] = df_films['duration_type'].replace('seasons','season')

# Eliminamos la columna 'duration' que ya no será útil debido a la creación de las dos anteriores
df_films = df_films.drop(['duration'], axis=1)
```

## Unión de los dos dataframes

Agregamos una nueva columna que llamaremos 'score' que formaremos a partir del promedio de las calificaciones (rating) de las calificaciones de los usuarios.

```python
# Creamos un dataframe con el valor promedio de los ratings (que llamaremos 'score') tomando como referncia el 'movieId'
df_score = df_ratings.groupby('movieId').mean().round(1).reset_index()
df_score = df_score.drop(['userId'], axis=1)  #eliminamos la columna 'userId' que no utilizaremos.

# Unimos el dataframe con el 'score' a la base de las peliculas con el método 'merge'ArithmeticError
df_films = df_films.merge(df_score, on='movieId', how='left')

# Comprobamos el cambio realizado
df_films.head()
```
